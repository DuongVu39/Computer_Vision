## Exploration Data Analysis

```python
labels = pd.read_csv("../data/dog_breed_data/labels.csv")
labels.head()
```

```python
print("There are {} numbers of dog breed in this data".format(labels.breed.nunique()))
```

```python
labels.breed.value_counts().head()
```

```python
labels.breed.value_counts().tail()
```

```python
sns.set_style("whitegrid")

plt.title("Number of each breed appeared in this dataset")
ax = labels.breed.value_counts().plot(kind='bar', figsize=(20,10), colormap='Pastel1')
ax.set(ylabel="Images per class", xlabel="Dog Breeds")
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
ax.title.set_size(20);
```
